```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
meta = pd.read_fwf('./ghcnm.tavg.v3.3.0.20160719.qca.inv',
                   names=['ID', 'LATITUDE', 'LONGITUDE', 'STNELEV', 'NAME', 'GRELEV', 'POPCLS', 'POPSIZ', 'TOPO', 'STVEG', 'STLOC', 'OCNDIS', 'AIRSTN', 'TOWNDIS', 'GRVEG', 'POPCSS'], 
                   widths=[12, 10, 10, 6, 29, 6, 1, 5, 2, 2, 2, 2, 1, 2, 16, 1],
                   header=None)
uk_ids = meta[ meta['NAME'].str.contains('\sUK$') ]['ID']
```

```python
tavg = pd.read_fwf('./ghcnm.tavg.v3.3.0.20160719.qca.dat', 
         colspecs=[(0,11),(11,15),(15,19),(19,24)], 
         names=['ID', 'YEAR', 'ELEMENT', 'VALUE1'], 
         header=None,
         na_values='-9999')
all = pd.merge(tavg, meta, on='ID', suffixes=['_l', '_r'])
```

```python
uk = all[ all['NAME'].str.contains('\sUK$') ]
uk
```

```python
uk_avgs_by_year = uk.groupby('YEAR').mean()['VALUE1'] / 100
uk_avgs_by_year
```

```python
uk_avgs_by_year.interpolate(method='cubic').plot()
```
